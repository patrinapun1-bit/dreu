```python

# Install Qiskit if needed:
# pip install qiskit qiskit-aer

from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram, plot_bloch_multivector
from qiskit.quantum_info import Statevector
import matplotlib.pyplot as plt
import numpy as np
print('Qiskit environment ready ✓')
```

```python

# A single qubit starts in |0⟩ by default
qc_zero = QuantumCircuit(1, 1)
qc_zero.measure(0, 0)
print('Circuit for |0⟩ state:')
print(qc_zero.draw(output='text'))

sim = AerSimulator()
job = sim.run(qc_zero, shots=1024)
counts = job.result().get_counts()
print(f'Measurement outcomes: {counts}')
```

```python

# The Hadamard gate (H) creates equal superposition:
# H|0⟩ = (|0⟩ + |1⟩) / √2

qc_super = QuantumCircuit(1, 1)
qc_super.h(0)          # Apply Hadamard
qc_super.measure(0, 0)
print('Circuit with Hadamard gate:')
print(qc_super.draw(output='text'))
```

```python

# Simulate and plot results
job = sim.run(qc_super, shots=2048)
counts = job.result().get_counts()
print(f'Measurement outcomes: {counts}')
plot_histogram(counts, title='Superposition Measurement (H|0⟩)')
plt.show()
```

```python

# Examine the statevector BEFORE measurement
qc_sv = QuantumCircuit(1)
qc_sv.h(0)

sv = Statevector.from_instruction(qc_sv)
print(f'Statevector: {sv}')
print(f'Prob |0⟩: {abs(sv[0])**2:.4f}')
print(f'Prob |1⟩: {abs(sv[1])**2:.4f}')
```

The second Hadamard gate reverses this operation, undoing the superposition. For example, if the qubit starts in the ∣0⟩ state, applying the Hadamard gate twice will return it to ∣0⟩.
