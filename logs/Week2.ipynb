```python
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
np.set_printoptions(precision=4, suppress=True)
print('NumPy ready ✓')
```

```python

# Computational basis states
ket_0 = np.array([[1], [0]], dtype=complex)   # |0⟩
ket_1 = np.array([[0], [1]], dtype=complex)   # |1⟩

print('|0⟩ ='); print(ket_0)
print('|1⟩ ='); print(ket_1)
```

```python

# Superposition state: |+⟩ = (|0⟩ + |1⟩) / √2
ket_plus  = (ket_0 + ket_1) / np.sqrt(2)
ket_minus = (ket_0 - ket_1) / np.sqrt(2)
print('|+⟩ ='); print(ket_plus)
print('|-⟩ ='); print(ket_minus)

# Verify normalization
print(f'\n‖|+⟩‖² = {np.linalg.norm(ket_plus)**2:.4f}  (should be 1.0)')
```

```python

# Standard single-qubit gates
I = np.eye(2, dtype=complex)
X = np.array([[0, 1], [1, 0]], dtype=complex)          # Pauli-X (NOT)
Y = np.array([[0, -1j], [1j, 0]], dtype=complex)       # Pauli-Y
Z = np.array([[1, 0], [0, -1]], dtype=complex)          # Pauli-Z
H = np.array([[1, 1], [1, -1]], dtype=complex) / np.sqrt(2)  # Hadamard

print('Hadamard matrix H:')
print(H)
print(f'\nH|0⟩ = '); print(H @ ket_0)
print(f'H|1⟩ = '); print(H @ ket_1)
```

```python

# Verify unitary: U†U = I
for name, gate in [('X',X),('Y',Y),('Z',Z),('H',H)]:
    product = gate.conj().T @ gate
    is_unitary = np.allclose(product, I)
    print(f'{name} is unitary: {is_unitary}')
```

```python
def inner(bra, ket):
    """Compute ⟨bra|ket⟩ (bra is already a ket, will be conjugate-transposed)"""
    return (bra.conj().T @ ket)[0, 0]

print(f'⟨0|0⟩ = {inner(ket_0, ket_0):.4f}  (should be 1)')
print(f'⟨0|1⟩ = {inner(ket_0, ket_1):.4f}  (should be 0 — orthogonal)')
print(f'⟨+|+⟩ = {inner(ket_plus, ket_plus):.4f}  (should be 1)')
print(f'⟨+|-⟩ = {inner(ket_plus, ket_minus):.4f}  (should be 0)')
```

```python
# Two-qubit state: |00⟩ = |0⟩ ⊗ |0⟩
ket_00 = np.kron(ket_0, ket_0)
ket_11 = np.kron(ket_1, ket_1)
print('|00⟩ ='); print(ket_00.T)
print('|11⟩ ='); print(ket_11.T)

# Bell state (maximally entangled): (|00⟩ + |11⟩)/√2
bell = (ket_00 + ket_11) / np.sqrt(2)
print('Bell state Φ⁺ ='); print(bell.T)
```

```python

# 2-qubit Hadamard on first qubit: H ⊗ I
HI = np.kron(H, I)
result = HI @ ket_00
print('(H⊗I)|00⟩ ='); print(result.T)
```

```python

import numpy as np

ket_0 = np.array([[1], [0]])
ket_1 = np.array([[0], [1]])

I = np.eye(2)

H = (1 / np.sqrt(2)) * np.array([
    [1, 1],
    [1, -1]
])

ket_01 = np.kron(ket_0, ket_1)
IH = np.kron(I, H)

result = IH @ ket_01
print(result.T)
```

```python
H_squared = H @ H
print(H_squared)
print(np.allclose(H_squared, I))
```
